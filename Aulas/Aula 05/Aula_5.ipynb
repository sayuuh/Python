```python
# Faça um jogo da sorte que guarde um número (0 - 99) e peça para o usuário tentar acertar o número, exiba se errou para mais ou menos
# 1. fácil -> 0 -> 25
# 2. Medio -> 0 -> 100
# 3. Dificil -> 0 -> 100 -> Cara -> Não dou a dica
# 4. ou > 4 -> Sair

from random import randint

continuar = True
segredo = randint(0, 100)
chances = 0
acertou = False

while(continuar):
  menu = input('**** JOGO DA SORTE **** \n 1 - Fácil \n 2 - Médio \n 3 - Difícil \n 4 - Sair \n R: ')
  if(menu.isnumeric()):
    menu = int(menu)
    if(menu > 0 and menu < 4):
      acertou = False
      chances = 0
      if(menu == 1):
        print('*** FÁCIL ***')
        segredo = randint(0, 25)
        while(not acertou): # == False
          chute = int(input('Digite um número de 0 a 25: ')) # pega o valor
          chances += 1 # tive mais de uma tentativa 
          if(chute == segredo):
            print('*** FIM DE PARTIDA ***')
            print('Você acertou!', segredo)
            print("Com {} tentativas. " .format(chances))
            acertou = True
          elif(chute > segredo):
            print('Menos')
          else:
            print('Mais')
      elif(menu == 2):
        print('*** MÉDIO ***')
        segredo = randint(0, 50)
        while(not acertou): # == False
          chute = int(input('Digite um número de 0 a 50: ')) # pega o valor
          chances += 1 # tive mais de uma tentativa 
          if(chute == segredo):
            print('*** FIM DE PARTIDA ***')
            print('Você acertou!', segredo)
            print("Com {} tentativas. " .format(chances))
            acertou = True
          elif(chute > segredo):
            print('Menos')
          else:
            print('Mais')
      else:
        print('*** Difícil ***')
        segredo = randint(0, 100)
        while(not acertou): # == False
          chute = int(input('Digite um número de 0 a 100: ')) # pega o valor
          chances += 1 # tive mais de uma tentativa 
          if(chute == segredo):
            print('*** FIM DE PARTIDA ***')
            print('Você acertou!', segredo)
            print("Com {} tentativas. " .format(chances))
            acertou = True
          else:
            sera = randint(0, 1) # cara/coroa par/impar sim/não
            if(sera == 0):
              if(chute > segredo):
                print('Menos')
              else:
                print('Mais')
            else: # se não for 0 não pode dar a dica
              print('NÃO POSSO TE AJUDAR')
    else:
      print('FIM DE JOGO')
      continuar = False
  else:
    print('Digite um valor válido!')

```
