LOCADORA
Criar um CRUD de uma locadora e guardar em um array

```python
'''
    C -> CRIAR
    R -> LER 
    U -> Atualizar
    D -> Deletar

    Filme -> (nome, diretor, ano)
'''
locadora = list()


def listarTodosOsFilmes():
    print('**********  LOCADORA DORA **********')
    print('__'* 30)

    for i in range(len(locadora)):
        print(f'{i} - {locadora[i]["nome"]}')

def cadastrarFilme():
    print('Cadastrar Filme.')
    filme = {}
    filme['nome'] = input('Digite o nome do filme: ')
    filme['diretor'] = input('Digite o nome do diretor do filme: ')
    filme['ano'] = input('Digite o ano do filme: ')
    locadora.append(filme.copy()) # Adicionando o dicionario a lista 
    print(f'{filme["nome"]} foi cadastrado com sucesso.')

def verificaExisteFilme():
    codigo = int(input('Digite o codigo do filme. R: '))
    for i in range(len(locadora)):
        if i == codigo:
            return True, codigo # Retornando uma lista -> [True, codigo]
    return False, 0
            
def descreverFilme():
    [filme, codigo] = verificaExisteFilme()
    if filme :
        print(f'Filme: {locadora[codigo]["nome"]}')
        print(f'Diretor: {locadora[codigo]["diretor"]}')
        print(f'Ano: {locadora[codigo]["ano"]}')
    else:
        print('Filme não foi encontrado.')

def deletarFilme():
    [filme, codigo] = verificaExisteFilme()
    if filme:
        print(f'Filme {locadora[codigo]["nome"]} foi deletado com sucesso')
        del locadora[codigo]
    else:
        print('Filme não foi encontrado.')

def alterarFilme():
    [filme, codigo] = verificaExisteFilme()
    if filme:
        locadora[codigo]['nome'] = input('Digite o nome do filme R: ')
        locadora[codigo]['diretor'] = input('Digite o nome do diretor do filme R: ')
        locadora[codigo]['ano'] = input('Digite o ano do filme R: ')
    else:
        print('Filme não encontrado')

while(True):
    menu = input(' 1 - Descrição do Filme \n 2 - Listar os Filme \n 3 - Cadastrar um Filme \n 4 - Alterar um Filme \n 5 - Deletar um Filme \n R: ')

    if menu == '1':
        descreverFilme()
    elif menu == '2':
        listarTodosOsFilmes()
    elif menu == '3':
        cadastrarFilme()
    elif menu == '4':
        alterarFilme() # Casa -> Alterar Nome, Diretor e o Ano do filme selecionado pelo usuario
    elif menu == '5':
        deletarFilme()
    else:
        print('Adeus ;(')
        break
```
